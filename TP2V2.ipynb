```python
numeric_columns = df.select_dtypes(include='number')

mean_numeric_values = df[numeric_columns.columns].mean()

df[numeric_columns.columns] = df[numeric_columns.columns].fillna(mean_numeric_values)

df.replace({False: 0, True: 1}, inplace=True)

df.replace({'die': 0, 'live': 1}, inplace=True)

print(df)
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(20,10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show() 

output_corr = correlation_matrix['class'].abs() > 0.5
malaise_corr = correlation_matrix['malaise'].abs() > 0.5

print('output_corr')
print(output_corr)
print('malaise_corr')
print(malaise_corr)

#selected_features1 = output_corr 

#selected_features2 = malaise_corr 

malaise_correlated_features = df.columns[malaise_corr ].drop("malaise")
df = df.drop(columns = malaise_correlated_features)

print("DataFrame after dropping highly correlated features with malaise:")
print(df.head(10))
```

```python
from sklearn.model_selection import train_test_split

target_column = 'class' 

X = df.drop(columns=[target_column])
y = df[target_column]

testsize = 0.2  

# Hold-Out Strategy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=testsize)

# Stratified Hold-Out Strategy
X_train_stratified, X_test_stratified, y_train_stratified, y_test_stratified = train_test_split(
    X, y, test_size=testsize,  stratify=y)

print("Hold-Out Strategy:")
print(f"Training Set Size: {X_train.shape[0]} samples")
print(f"Test Set Size: {X_test.shape[0]} samples")

print("\nStratified Hold-Out Strategy:")
print(f"Training Set Size: {X_train_stratified.shape[0]} samples")
print(f"Test Set Size: {X_test_stratified.shape[0]} samples")
print("\n")


def display_class_samples(y_train, y_test, strategy_name):
    print(f"{strategy_name}:")
    for class_label in set(y_train):
        train_samples = (y_train == class_label).sum()
        test_samples = (y_test == class_label).sum()
        print(f"Class {class_label}: Train samples={train_samples}, Test samples={test_samples}")
        print("\n")
display_class_samples(y_train, y_test, "Hold-Out Strategy")
display_class_samples(y_train_stratified, y_test_stratified, "Stratified Hold-Out Strategy")
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

X = df.drop('Class', axis=1) 
y = df['Class']

# Encode the target labels to integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Define preprocessing for numeric and categorical features
numeric_features = ['deg-malig']
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))  
])

categorical_features = ['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'breast', 'breast-quad', 'irradiat']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine numeric and categorical transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create the pipeline with preprocessing and the KNN classifier
knn_classifier = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', KNeighborsClassifier(n_neighbors=3))
])
knn_classifier.fit(X, y)
# Perform 5-fold cross-validation 
cv_results = cross_val_predict(knn_classifier, X, y_encoded, cv=5, method='predict_proba')

# Display confusion matrix for each split
for i in range(5):
    y_pred_cv = (cv_results[:, 1] > 0.5).astype(int)  # The resulting y_pred_cv is a binary array of predicted class labels.
    cm = confusion_matrix(y_encoded, y_pred_cv)
    print(f'Confusion Matrix - Split {i + 1}:\n{cm}')

# Calculate and display mean recall, precision, and f1-score
report = classification_report(y_encoded, y_pred_cv, target_names=label_encoder.classes_)
print(f'Classification Report:\n{report}')

# Plot ROC curves for each fold independently
fig, axes = plt.subplots(1, 5, figsize=(15, 3), sharex=True, sharey=True)
for i, ax in enumerate(axes):
    # Perform 5-fold cross-validation
    cv_results = cross_val_predict(knn_classifier, X, y_encoded, cv=5, method='predict_proba')

    fpr, tpr, _ = roc_curve(y_encoded, cv_results[:, 1])  # Assuming binary classification
    roc_auc = auc(fpr, tpr)

    ax.plot(fpr, tpr, lw=2, label=f'Fold {i + 1} (AUC = {roc_auc:.2f})')
    ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='black', label='Random', alpha=0.8)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve - Fold {i + 1}')
    ax.legend(loc='lower right')

plt.suptitle('Receiver Operating Characteristic (ROC) Curves for Each Fold', y=1.05)
plt.tight_layout()
plt.show()
```

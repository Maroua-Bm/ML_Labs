```python
class_distribution = breast_cancer['Class'].value_counts()

# Plotting the distribution
plt.figure(figsize=(8, 6))
class_distribution.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Samples According to Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
### Class Imbalance:there's a significant difference in the number of samples between classes. 
###it indicates a class imbalance, which might affect the performance of models.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Assuming 'breast_cancer' is your DataFrame
# 'Class' is the target variable, and you have multiple categorical features

# Copy the DataFrame to avoid modifying the original data
df_numeric = breast_cancer.copy()

# Convert categorical columns to numeric using LabelEncoder
label_encoder = LabelEncoder()

categorical_columns = ['menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat']

for column in categorical_columns:
    df_numeric[column] = label_encoder.fit_transform(df_numeric[column])

# Scatter plots for pairwise feature distributions colored by class
sns.pairplot(df_numeric, hue='Class', vars=categorical_columns, markers=["o", "s"])
plt.show()
```

```python
#Check for Gaussian Distribution 
from scipy import stats

features_to_check = ['menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat']

# Plot histograms for each feature by class
for feature in features_to_check:
    plt.figure(figsize=(8, 6))
    for class_label in breast_cancer['Class'].unique():
        sns.histplot(breast_cancer[breast_cancer['Class'] == class_label][feature], label=f'Class {class_label}')
    plt.title(f'Histogram of {feature} by Class')
    plt.xlabel(feature)
    plt.legend()
    plt.show()
```

```python
#logistic regression 
# Select features to plot
features_to_plot = ['menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat']

# Add 'Class' to the selected features
features_to_plot.append('Class')

# Subset the DataFrame with selected features
df_subset = breast_cancer[features_to_plot]

# Plot pairwise scatter plots
sns.pairplot(df_subset, hue='Class', markers=["o", "s"], palette='viridis')
plt.show()
```

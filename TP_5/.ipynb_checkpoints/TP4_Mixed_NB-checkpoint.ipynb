```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, normaltest


# Perform Shapiro-Wilk test for normality on each numerical column
def check_normality(data):
    for column in data.select_dtypes(include=[np.number]).columns:
        _, p_value_shapiro = shapiro(data[column])
        _, p_value_normaltest = normaltest(data[column])
        
        print(f"Shapiro-Wilk test for {column}: p-value = {p_value_shapiro}")
        print(f"Normality test for {column}: p-value = {p_value_normaltest}")
        
        if p_value_shapiro > 0.05 and p_value_normaltest > 0.05:
            print(f"The data for {column} appears to be normally distributed.\n")
        else:
            print(f"The data for {column} does not appear to be normally distributed.\n")
        
        # Plot histogram for visual inspection
        plt.figure(figsize=(8, 4))
        sns.histplot(data[column], kde=True, color='skyblue')
        plt.title(f"Histogram for {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()

check_normality(breast_cancer)

# Check for independence between features
def check_feature_independence(data):
    numeric_columns = data.select_dtypes(include=[np.number]).columns
    sns.heatmap(data[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Correlation Heatmap")
    plt.show()

check_feature_independence(breast_cancer)
```

```python
def calculate_class_probs(df, target_column='Class'):
    class_probs = df[target_column].value_counts(normalize=True).to_dict()
    return class_probs

# Sample Usage
# Assuming 'df' is your DataFrame
class_probs = calculate_class_probs(breast_cancer)

# Print the calculated probabilities for each unique class value
for class_value, prob in class_probs.items():
    print(f"Class: {class_value}, Probability: {prob}")
```

```python
import pandas as pd

def calculate_mean_std_numerical(df, target_column='Class'):
    numerical_columns = df.select_dtypes(include=[np.number]).columns
    class_values = df[target_column].unique()

    mean_std_by_class = {}

    for class_value in class_values:
        class_data = df[df[target_column] == class_value]
        mean_std_by_class[class_value] = class_data[numerical_columns].agg([np.mean, np.std])

    return mean_std_by_class

# Sample Usage
df = pd.read_csv('breast-cancer_csv.csv')  # Replace 'your_dataset.csv' with your actual file
mean_std_numerical = calculate_mean_std_numerical(df)

# Print mean and standard deviation for each numerical column for each class
for class_value, stats in mean_std_numerical.items():
    print(f"Class: {class_value}")
    print(stats)
    print("\n")
```

```python
def calculate_prob_cat(df, target_column='Class'):
    categorical_columns = df.select_dtypes(exclude=[np.number]).columns
    class_values = df[target_column].unique()

    prob_cat_by_class = {}

    for class_value in class_values:
        class_data = df[df[target_column] == class_value]
        prob_cat_by_class[class_value] = {}

        for column in categorical_columns:
            if column != target_column:
                value_counts = class_data[column].value_counts(normalize=True).to_dict()
                prob_cat_by_class[class_value][column] = value_counts

    return prob_cat_by_class

# Sample Usage
prob_cat = calculate_prob_cat(df)

# Print probabilities for each unique value in categorical columns for each class
for class_value, prob_dict in prob_cat.items():
    print(f"Class: {class_value}")
    for column, values_prob in prob_dict.items():
        print(f"{column}:")
        print(values_prob)
        print("\n")
```

```python
def calculate_numerical_prob(sample, mean_std_numerical, class_values, numerical_columns):
    numerical_probs = {}

    for class_value in class_values:
        class_probs = 1.0

        for column in numerical_columns:
            mean = mean_std_numerical[class_value].loc['mean', column]
            std = mean_std_numerical[class_value].loc['std', column]
            value = sample[column]

            # Calculate probability using Gaussian (Normal) Distribution
            class_probs *= (1 / (std * (2 * 3.14159) ** 0.5)) * \
                           2.71828 ** (-0.5 * ((value - mean) / std) ** 2)

        numerical_probs[class_value] = class_probs

    return numerical_probs

# Sample Usage
# Assuming 'sample_data' is a dictionary containing values for numerical columns
sample_data = {'age': '40-49', 'menopause': 'premeno', 'tumor-size': '35-39', 'inv-nodes': '0-2',
                     'node-caps': 'yes', 'deg-malig': 3, 'breast': 'right', 'breast-quad': 'left_low',
                     'irradiat': 'yes', 'Class': 'no-recurrence-events'}

# Assuming 'mean_std_numerical' is the result from the previous step
numerical_columns = df.select_dtypes(include=[np.number]).columns
numerical_probs = calculate_numerical_prob(sample_data, mean_std_numerical, df['Class'].unique(), numerical_columns)

# Print the calculated probabilities for each class with corresponding attributes
for class_value, prob in numerical_probs.items():
    print(f"Class: {class_value}, Probability: {prob} for {', '.join(numerical_columns)}")
```

```python
def calculate_categorical_prob(sample, prob_cat, class_values):
    categorical_probs = {}

    for class_value in class_values:
        class_probs = {}

        for column, value in sample.items():
            if column != 'Class' and column in prob_cat[class_value]:
                class_probs[column] = prob_cat[class_value][column].get(value, 1e-5)

        categorical_probs[class_value] = class_probs

    return categorical_probs

# Sample Usage
# Assuming 'sample_data' is a dictionary containing values for categorical columns
sample_data_categorical = {'age': '40-49', 'menopause': 'premeno', 'tumor-size': '35-39', 'inv-nodes': '0-2',
                            'node-caps': 'yes', 'deg-malig': 3, 'breast': 'right', 'breast-quad': 'left_low',
                            'irradiat': 'yes', 'Class': 'no-recurrence-events'}

# Assuming 'prob_cat' is the result from the previous step
categorical_probs = calculate_categorical_prob(sample_data_categorical, prob_cat, df['Class'].unique())

# Print the calculated probabilities for each class with corresponding attributes
for class_value, probs_dict in categorical_probs.items():
    print(f"Class: {class_value}")
    for column, prob in probs_dict.items():
        print(f"  {column}: {prob}")
    print("\n")
```

```python
def predict_probabilities(sample, mean_std_numerical, prob_cat, class_probs, class_values):
    numerical_probs = calculate_numerical_prob(sample, mean_std_numerical, class_values, numerical_columns)
    categorical_probs = calculate_categorical_prob(sample, prob_cat, class_values)
    
    final_probs = {}

    for class_value in class_values:
        # Multiply numerical and categorical probabilities
        prob_numerical = numerical_probs[class_value]
        prob_categorical = np.prod(list(categorical_probs[class_value].values()))
        prob_class = class_probs[class_value]

        # Calculate final probability
        final_prob = prob_numerical * prob_categorical * prob_class
        final_probs[class_value] = final_prob

    # Normalize probabilities to sum to 1
    sum_probs = sum(final_probs.values())
    normalized_probs = {class_value: prob / sum_probs for class_value, prob in final_probs.items()}

    predicted_class = max(final_probs, key=final_probs.get)
    print(f"\nPredicted Class: {predicted_class}")

    return normalized_probs

# Sample Usage
# Assuming 'sample_data' is a dictionary containing values for all columns
sample_data = {'age': '40-49', 'menopause': 'premeno', 'tumor-size': '35-39', 'inv-nodes': '0-2',
               'node-caps': 'yes', 'deg-malig': 3, 'breast': 'right', 'breast-quad': 'left_low',
               'irradiat': 'yes', 'Class': 'no-recurrence-events'}

# Assuming 'mean_std_numerical', 'prob_cat', and 'class_probs' are results from previous steps
numerical_columns = df.select_dtypes(include=[np.number]).columns
class_values = df['Class'].unique()

final_probs = predict_probabilities(sample_data, mean_std_numerical, prob_cat, class_probs, class_values)

# Print the final probabilities for each class value
for class_value, prob in final_probs.items():
    print(f"Class: {class_value}, Probability: {prob}")
```

```python
def display_model_parameters(mean_std_numerical, prob_cat, class_probs):
    print("Mean and Standard Deviation for Numerical Columns:")
    for class_value, stats in mean_std_numerical.items():
        print(f"\nClass: {class_value}")
        print(stats)

    print("\nProbabilities for Categorical Columns:")
    for class_value, prob_dict in prob_cat.items():
        print(f"\nClass: {class_value}")
        for column, values_prob in prob_dict.items():
            print(f"{column}:")
            print(values_prob)

    print("\nClass Probabilities:")
    for class_value, prob in class_probs.items():
        print(f"Class: {class_value}, Probability: {prob}")

# Example Usage
display_model_parameters(mean_std_numerical, prob_cat, class_probs)
```

```python
def export_model_parameters(mean_std_numerical, prob_cat, class_probs, export_path='model_parameters.csv'):
    # Convert model parameters to a DataFrame for easier export
    df_mean_std = pd.concat([v.T for v in mean_std_numerical.values()], keys=mean_std_numerical.keys())
    df_prob_cat = pd.concat([pd.DataFrame(v) for v in prob_cat.values()], keys=prob_cat.keys())
    df_class_probs = pd.DataFrame(list(class_probs.items()), columns=['Class', 'Probability'])

    # Save DataFrames to CSV (you can choose other formats as well)
    df_mean_std.to_csv(export_path.replace('.csv', '_mean_std.csv'))
    df_prob_cat.to_csv(export_path.replace('.csv', '_prob_cat.csv'))
    df_class_probs.to_csv(export_path.replace('.csv', '_class_probs.csv'))


export_model_parameters(mean_std_numerical, prob_cat, class_probs, export_path='Mixed_NB_Parameters.csv')
```

```python
def import_model_parameters(import_path='model_parameters.csv'):
    mean_std_numerical = {}
    prob_cat = {}
    class_probs = {}

    # Load DataFrames from CSV
    df_mean_std = pd.read_csv(import_path.replace('.csv', '_mean_std.csv'), header=[0, 1], index_col=0)
    df_prob_cat = pd.read_csv(import_path.replace('.csv', '_prob_cat.csv'), header=[0, 1], index_col=0)
    df_class_probs = pd.read_csv(import_path.replace('.csv', '_class_probs.csv'), index_col=0)

    # Convert DataFrames back to dictionaries
    for class_value in df_mean_std.columns.get_level_values(0).unique():
        mean_std_numerical[class_value] = df_mean_std[class_value].T

    for class_value in df_prob_cat.columns.get_level_values(0).unique():
        prob_cat[class_value] = df_prob_cat[class_value].to_dict()

    class_probs = df_class_probs['Probability'].to_dict()

    return mean_std_numerical, prob_cat, class_probs

# Example Usage
imported_mean_std, imported_prob_cat, imported_class_probs = import_model_parameters(import_path='Mixed_NB_Parameters.csv')
```

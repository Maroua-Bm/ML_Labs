### Dataset training and evaluation

#### Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
x_train, x_test,y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=0) 
```

<div class="alert alert-block alert-danger">
    
5. Train the data using the LinearRegression module from sklearn.6. 
Display the model's equation
   .7. 
Create a plot showing the data and the generated model.
<div/>

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
plt.scatter(x_test, y_test, color='black')
plt.plot(x_test, y_pred, color='blue', linewidth=3)
plt.title('Linear Regression Model')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
# slope: B1 : the coefficient associated with the independent variable"hours" 
#intercept : B0: 
slope = model.coef_[0]
intercept = model.intercept_
print(f'Model Equation: Scores = {slope:.2f} * Hours + {intercept:.2f}')
```

<div class="alert alert-block alert-danger">
    
8. Print the performance of the generated model on the training and the test set, based on the following metrics: MAE, MSE, RMSE, and R square. Does the generated model present an  overfitting problem?
9. Describe the R square metric and its advantages compared to the other metrics.
10. Is the generated model characterized by a good performance?
11. Create a plot visualizing the difference between actual and predicted values (R-square).

<div/>

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Make predictions on both training and test sets
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

# Calculate performance metrics
def calculate_metrics(y_orig, y_pred):
    mae = mean_absolute_error(y_orig, y_pred)
    mse = mean_squared_error(y_orig, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_orig, y_pred)
    return mae, mse, rmse, r2

# Calculate metrics for training set
mae_train, mse_train, rmse_train, r2_train = calculate_metrics(y_train, y_train_pred)

# Calculate metrics for test set
mae_test, mse_test, rmse_test, r2_test = calculate_metrics(y_test, y_test_pred)

# Display the performance metrics
print("Training Set Metrics:")
print(f"MAE: {mae_train:.2f}")
print(f"MSE: {mse_train:.2f}")
print(f"RMSE: {rmse_train:.2f}")
print(f"R-squared: {r2_train:.2f}")
print("\nTest Set Metrics:")
print(f"MAE: {mae_test:.2f}")
print(f"MSE: {mse_test:.2f}")
print(f"RMSE: {rmse_test:.2f}")
print(f"R-squared: {r2_test:.2f}")

# Check for overfitting
if r2_train > r2_test:
    print("\nThe model may be overfitting as R-squared on the training set is higher than on the test set.")
else:
    print("\nThe model does not show clear signs of overfitting.")
```

```python
'''R-squared, or the coefficient of determination, measures the proportion of the variance in the dependent variable (target) 
that is explained by the independent variable(s) (features) in the model. It is a value between 0 and 1, where 0 indicates that 
the model does not explain any variance, and 1 indicates a perfect fit.'''
```

```python
'''Advantages of R-squared:
Easy Interpretation: R-squared is intuitive and easy to interpret. It provides a percentage indicating how well the model explains 
the variability in the data.Scale Independence: R-squared is scale-independent, meaning it doesn't depend on the scale of the variables. 
This allows for easier comparison between different models.Useful for Model Comparison: R-squared can be used to compare different models. 
A higher R-squared generally indicates a better fit, although it should be considered alongside other metrics.'''
```

```python
'''Whether the model is considered to have good performance depends on the specific requirements and context. 
Generally, a higher R-squared, lower MAE, MSE, and RMSE are indicative of better performance. 
However, it's essential to consider the nature of the problem and the acceptable level of error.'''
```

```python
# Scatter plot for the training set
plt.scatter(x_train, y_train, color='blue', label='Actual')
plt.scatter(x_train, y_train_pred, color='red', label='Predicted')
plt.title('Actual vs Predicted (Training Set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()

# Scatter plot for the test set
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.scatter(x_test, y_test_pred, color='red', label='Predicted')
plt.title('Actual vs Predicted (Test Set)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

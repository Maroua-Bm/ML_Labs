### Dataset training and evaluation

#### Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
x_train, x_test,y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_train)
```

```python
#EQUATION
slope = model.coef_[0] #B1 : the coefficient associated with the independent variable"hours" 
intercept = model.intercept_ #B0
equation = f'y = {slope}x + {intercept}'
print(f'Model Equation: Scores = {slope:.2f} * Hours + {intercept:.2f}')
```

```python
plt.scatter(x_train, y_train, alpha=0.7, label='Data Points')
plt.plot(x_train, model.predict(x_train), color='red', label='Linear Regression Model')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.show()
```

<div class="alert alert-block alert-danger">
    
8. Print the performance of the generated model on the training and the test set, based on the following metrics: MAE, MSE, RMSE, and R square. Does the generated model present an  overfitting problem?
9. Describe the R square metric and its advantages compared to the other metrics.
10. Is the generated model characterized by a good performance?
11. Create a plot visualizing the difference between actual and predicted values.

<div/>

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_train_predicted = model.predict(x_train)
y_test_predicted = model.predict(x_test)

mae_train = mean_absolute_error(y_train, y_train_predicted)
mae_test = mean_absolute_error(y_test, y_test_predicted)

mse_train = mean_squared_error(y_train, y_train_predicted)
mse_test = mean_squared_error(y_test, y_test_predicted)

rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

r2_train = r2_score(y_train, y_train_predicted)
r2_test = r2_score(y_test, y_test_predicted)


print("Performance Metrics:")
print(f"MAE Of Training: {mae_train:.2f}")
print(f"MAE Of Test: {mae_test:.2f}")
print(f"MSE Of Training: {mse_train:.2f}")
print(f"MSE Of Test: {mse_test:.2f}")
print(f"RMSE Of Training: {rmse_train:.2f}")
print(f"RMSE Of Test: {rmse_test:.2f}")
print(f"R^2 Of Training: {r2_train:.2f}")
print(f"R^2 Of Test: {r2_test:.2f}")


if r2_train > r2_test:
    print("The model may have overfitting, as R^2 is higher on the training set.")
else:
    print("The model does not seem to have overfitting.")

```

<div>

R-square :
- Definition::R^2 involves the difference between predicted and actual values.
- Range: It's a number between 0 and 1.
- Interpretation:
  - R2=0 The model doesn't help us predict anythingthing
  - R2=1 The model perfectly predicts what's happening
- Comparison: This comparison between R_train and R_test helps ensure that the model is not just memorizing the training data but also works well on nAdvantages of R-squared:
- Easy Interpretation: R-squared is intuitive and easy to interpret. It provides a percentage indicating how well the model explains 
the variability in the data.
- Scale Independence: R-squared is scale-independent, meaning it doesn't depend on the scale of the variables. 
This allows for easier comparison between different models.
-Useful for Model Comparison: R-squared can be used to compare different models. 
A higher R-squared generally indicates a better fit, although it should be considered alongside other metrics.

Generally, a higher R-squared, lower MAE, MSE, and RMSE are indicative of better performance.
So, our model have a good performance.ew data.

```python
from sklearn.model_selection import train_test_split

x = normalized_df[['RM', 'PTRATIO', 'LSTAT']]
y = normalized_df['MEDV']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

4. SGD is an iterative stochastic optimization technique. Print the convergence graph of the trained model ( x presents the iterations and y is the loss).
<div/>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error

learning_rates = [0.01, 0.001, 0.0001]
colors = ['red', 'green', 'blue']

plt.figure(figsize=(10, 6))

for lr, color in zip(learning_rates, colors):
    sgdr = SGDRegressor(learning_rate='invscaling', eta0=lr, max_iter=1000, verbose=0)  

    train_losses = []
    test_losses = []

    for epoch in range(1000):
        sgdr.partial_fit(x_train, y_train)

        # the training set
        y_train_pred = sgdr.predict(x_train)
        train_loss = mean_squared_error(y_train, y_train_pred)
        train_losses.append(train_loss)

        # the test set
        y_test_pred = sgdr.predict(x_test)
        test_loss = mean_squared_error(y_test, y_test_pred)
        test_losses.append(test_loss)

    #  the convergence graphs
    plt.plot(np.arange(len(train_losses)), train_losses, label=f'Training Loss (LR={lr})', color=color, linestyle='dashed')
    plt.plot(np.arange(len(test_losses)), test_losses, label=f'Test Loss (LR={lr})', color=color)

plt.title('Convergence of SGDRegressor for Different Learning Rates')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
